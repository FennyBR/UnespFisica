8. O IPVA no estado de São Paulo consiste em 4% do valor do carro, entretanto, caso

de 3%. Faça um programa que receba o valor do carro e a categoria que o

pago.

```python
val_carro = cod = ipva = 0

val_carro = float(input('Digite o valor do carro: '))
cod = int(input('Digite: \n1 se for PCD \n2 se exerce atividade remunerada \n3 se não se encaixa em nenhuma das afirmações \n'))

if cod == 1:
  ipva = val_carro * 0.02
elif cod == 2:
  ipva = val_carro * 0.03
else:
  ipva = val_carro * 0.04

print('O valor do IPVA a ser pago é R$', ipva)
```

caminhão é necessário saber a quantidade de eixos (de 2 a 9) e multiplicar pela

```python
veiculo = eixo = taxa = 0

veiculo = int(input('Digite: \n1 para carro \n2 para caminhão \n3 para moto \n'))

if veiculo == 2:
  eixo = int(input('Digite o número de eixos do caminhão: '))

  if eixo > 9 or eixo < 2:
    print('Valor não aceito!')
  else:
    taxa = eixo * 7.84
elif veiculo == 1:
  taxa = 7.84
else:
  taxa = 7.84/3

print('O valor a ser pago no pedágio é R$', round(taxa, 2))
```
